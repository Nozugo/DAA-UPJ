## Shortest Path

- Path adalah urutan node antara node awal dan node akhir, di mana tidak ada node yang muncul dua kali pada jalur tersebut. Jalur mewakili rute antara titik awal dan akhir yang dipilih. Ini akan menjadi himpunan simpul, p, menghubungkan simpul awal dengan simpul akhir. Tidak ada simpul yang diulang di hal.

- *Length of path* dihitung dengan menghitung sisi-sisi penyusunnya. Dari semua opsi, jalur dengan panjang terkecil disebut jalur terpendek. Penghitungan jalur terpendek digunakan secara luas dalam algoritme teori graf, tetapi tidak selalu mudah untuk dihitung. Ada berbagai algoritme yang dapat digunakan untuk menemukan jalur terpendek antara node awal dan node akhir. Salah satu algoritme paling populer yang dapat digunakan untuk menemukan jarak terpendek antara simpul awal dan simpul akhir adalah algoritme Dijkstra, yang diterbitkan pada akhir 1950-an. Ini menghitung jalur terpendek dalam grafik.

- Ini dapat digunakan oleh perangkat Global Positioning System (GPS) untuk menghitung jarak minimum antara sumber dan tujuan. Algoritme Dijkstra juga digunakan dalam algoritme perutean jaringan.

```python
import networkx as nx
import matplotlib.pyplot as plt

vertices = range(1, 10)
edges = [(7, 2), (2, 3), (7, 4), (4, 5), (7, 3), (7, 5), (1, 6), (1, 7), (2, 8), (2, 9)]
G = nx.Graph()
G.add_nodes_from(vertices)
G.add_edges_from(edges)
nx.draw(G, with_labels=True, node_color='y', node_size=800)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
centrality = nx.eigenvector_centrality(G)
sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
```

```python
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20, 20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size)
plt.axis('off')
```
