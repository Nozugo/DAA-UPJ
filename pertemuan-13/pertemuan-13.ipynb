## Algoritma Breadth First Search

Algoritma Breadth-first Search (BFS) adalah algoritma pencarian yang dimulai dari simpul awal dan bergerak melebar, mengunjungi semua simpul yang bertetangga dengan simpul saat ini terlebih dahulu sebelum mengunjungi simpul yang lebih jauh.

```python
def bfs(graph, start):
    visited = []
    queue = []
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            neighbours = graph[node]
            for neighbour in neighbours:
                queue.append(neighbour)
    return visited
```

```python
graph = {
    'Amin': ['Mike', 'Nick', 'Wasim'],
    'Wasim': ['Imran', 'Amin'],
    'Imran': ['Wasim', 'Faras'],
    'Faras': ['Imran'],
    'Mike': ['Amin'],
    'Nick': ['Amin']
}
```

## Algoritma Depth First Search

DFS (Depth-First Search) merupakan alternatif dari BFS (Breadth-First Search) yang digunakan untuk mencari data dari sebuah graf.

Perbedaan utama antara DFS dan BFS adalah bahwa setelah memulai dari simpul akar (root vertex), algoritma DFS akan menelusuri simpul sedalam mungkin pada setiap jalur tunggal yang unik secara berurutan.

Untuk setiap jalur, setelah mencapai kedalaman maksimal, algoritma akan menandai semua simpul yang terkait dengan jalur tersebut sebagai telah dikunjungi.

## Enqueue dan Dequeue

- **Enqueue** adalah operasi untuk menambahkan data ke ujung belakang queue. Operasi ini biasanya dilakukan dengan menggunakan fungsi append().

- **Dequeue** adalah operasi untuk menghapus data dari ujung depan queue. Operasi ini biasanya dilakukan dengan menggunakan fungsi pop().

```python
def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
```

```python
graph = {
    'Amin': ['Mike', 'Nick', 'Wasim'],
    'Wasim': ['Imran', 'Amin'],
    'Imran': ['Wasim', 'Faras'],
    'Faras': ['Imran'],
    'Mike': ['Amin'],
    'Nick': ['Amin']
}

dfs(set(), graph, 'Amin')
```
