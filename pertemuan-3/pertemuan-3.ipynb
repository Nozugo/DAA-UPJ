## Lists

```python
# List merupakan salah satu tipe data yang dapat diubah dan mempunyai index

bin_colors = ['Red', 'Green', 'Blue', 'Yellow']
print(bin_colors[1]) # Green

# [ ---- Slicing  ---- ]

print(bin_colors[0:3]) # ['Red', 'Green']
print(bin_colors[:3]) # ['Red', 'Green', 'Blue']
print(bin_colors[::1]) # ['Red', 'Green', 'Blue', 'Yellow']

print(bin_colors[2:0:-1]) # ['Blue', 'Green']

for i in bin_colors:
    print(i, "Square")
    """
    Ouput:
        Red Square
        Green Square
        Blue Square
        Yellow Square
    """
```

## Tuples

```python
# Tuples merupakan salah satu tipe data yang memiliki index, sama seperti lists tetapi tidak bisa diubah elementnya.

bin_colors = ('Red', 'Green','Blue','Yellow')
print(bin_colors[1]) # Green
print(bin_colors[2:]) # ('Blue', 'Yellow')

hariAwal = ("senin","selasa","rabu")
hariAkhir = ("kamis","jumat","sabtu")

hari = (hariAwal, hariAkhir)
print(hari) # (('senin', 'selasa', 'rabu'), ('kamis', 'jumat', 'sabtu'))

# Tuples dapat berisi berbagai macam tipe data
acak = (True, {'name' : 'depotkode','status': 'alive'}, [1,2,3])
print(acak) # (True, {'name': 'depotkode', 'status': 'alive'}, [1, 2, 3])
```

## Dictionary

```python
# Berbeda dengan tuples dan lists, tipe data dictionary mempunyai key dan value. Selain itu, key di dalam dictionary tidak boleh duplicate.

bin_colors = {
  "manual_color": "Yellow",
  "approved_color": "Green",
  "refused_color": "Red"
}

print(bin_colors) # {'manual_color': 'Yellow', 'approved_color': 'Green', 'refused_color': 'Red'}

print(bin_colors.get("manual_color")) # Yellow

bin_colors["approved_color"] = "Purple"
print(bin_colors) # {'manual_color': 'Yellow', 'approved_color': 'Purple', 'refused_color': 'Red'}
```

## Sets

```python
# Tipe data sets digunakan untuk menyimpan berbagai macam item dalam satu variable. Element di dalamnya tidak dapat diubah tetapi bisa ditambah dan dihapus, tidak boleh duplicate, dan tidak teratur sehingga tipe data ini tidak memiliki index.

green = {'grass', 'leaves'}
print(green) # {'leaves', 'grass'}

# list menjadi set
set_01 = set([1, 2, 3, 4])
print(type(set_01)) # <class 'set'>
print(set_01) # {1, 2, 3, 4}

set_01.add(0)
print(set_01) # {0, 1, 2, 3, 4}
set_01.discard(1)
print(set_01) # {0, 2, 3, 4}

# Union Set
set_A = {1,2,3,4}
set_B = {3,4,5,6}

print(set_A|set_B) # {1, 2, 3, 4, 5, 6}
print(set_A.union(set_B)) # {1, 2, 3, 4, 5, 6}

# Intersection Set
print(set_A & set_B) # {3, 4}
print(set_A.intersection(set_B)) # {3, 4}

# Difference Set
# A - B
print(set_A - set_B) # {1, 2}
print(set_A.difference(set_B)) # {1, 2}

# B - A
print(set_B - set_A) # {5, 6}
print(set_B.difference(set_A)) # {5, 6}

# Symmetric Difference
print(set_A ^ set_B) # {1, 2, 5, 6}
print(set_A.symmetric_difference(set_B)) # {1, 2, 5, 6}

```

## Matrix

```python
# Matriks adalah struktur data dua dimensi dengan jumlah kolom dan baris yang tetap.

import numpy as np
myMatrix = np.array([[11, 12, 13], [21, 22, 23], [31, 32, 33]])
print(myMatrix)
"""
Output:

[[11 12 13]
 [21 22 23]
 [31 32 33]]
"""
print(type(myMatrix)) # <class 'numpy.ndarray'>

# Transpose matrix
matriksC = np.array([[1,2,3],
		                [4,5,6]])
print("Matriks C : ")
print(matriksC)

transpose_matriks = np.transpose(matriksC)

print("Tranpose Matriks C : ")
print(transpose_matriks)

"""
Output:

Matriks C :
[[1 2 3]
 [4 5 6]]
Tranpose Matriks C :
[[1 4]
 [2 5]
 [3 6]]
"""
```
